```python
from py2neo import Graph,Node,Relationship
import pandas as pd
import matplotlib.pyplot as plt
```

```python
graph = Graph("bolt://localhost:7687", auth=('neo4j', '123456789'))
```

```python
person=Node("Peson",name="John",age=30)
```

```python
graph.create(person)
```

```python
#Cypher Query
number_of_person_nodes="MATCH(p:COURSE) RETURN Count(p)"
number_of_movie_nodes="MATCH(m:Student) RETURN Count(m)"
#Evaluate the Cypher query
result_persons=graph.evaluate(number_of_person_nodes)
result_movies=graph.evaluate(number_of_movie_nodes)
#Print the result
print(f"No of person node is {result_persons} & No of movie node is {result_movies}")
```

```python
neoud="MATCH(n) RETURN n"
graph.evaluate(neoud)
```

```python
df_result_count=pd.Series({'Person':result_persons,'Movie':result_movies})
df_result_count.plot(kind='bar',color=['blue','darkorange'])
plt.xlabel('Node Label')
plt.ylabel('No Of Nodes')
```

```python
cypher_all_relationship="MATCH (n)-[r]-(m) RETURN DISTINCT type(r) as RelationType"
graph.run(cypher_all_relationship).to_data_frame()
```

```python
#Cypher to fetch the actors along with their movie
all_actors_name="MATCH (p:Student)-[rel:ENROLLED_IN]->(m:COURSE) RETURN DISTINCT p.name as StudentName,m.name as CourseName ORDER BY StudentName ASC"
#Run the above Cypher
graph.run(all_actors_name).to_data_frame()
```

```python
#Create a new node for actor
new_person = Node("Person",name="Aamir Khan",born="1965")
new_person
```

```python
#Create a new node for movie
new_student=Node("Student",name="Houssam",id=15,age=20)
new_student
```

```python
#Now insert both the nodes in the graph
tx=graph.begin() #Get the transaction object which will perform this transaction in Graph
tx.create(new_person)
tx.create(new_student)
#Create the relationship between above two nodes
tx.create(Relationship(new_person, "ACTED_IN", new_student))
tx.commit() #Commit the result in the Graph
```

```python
#Cypher to fetch the actors along with their movie
all_actors_name="MATCH (p:Person)-[rel:ACTED_IN]->(m:Student) RETURN DISTINCT p.name as PersonName,m.name as StudentName ORDER BY PersonName ASC"
#Run the Cypher
graph.run(all_actors_name).to_data_frame()
```

```python
# Importation des bibliothèques nécessaires
from py2neo import Graph
import osmnx as ox
import networkx as nx

# Connexion à la base de données Neo4j
graph = Graph("bolt://localhost:7687", auth=("neo4j", "password"))

# Téléchargement et traitement des données OpenStreetMap
place_name = "Kamppi, Helsinki, Finland"
graph_data = ox.graph_from_place(place_name, network_type='drive')

# Conversion des données en graph Neo4j
G = ox.neo4j.graph_to_gdfs(graph_data, nodes=True, edges=True, node_geometry=True, fill_edge_geometry=True)

# Création des nœuds et des relations dans Neo4j
for _, row in G.nodes.iterrows():
    graph.run("CREATE (n:Node {osmid: $osmid, y: $y, x: $x})", osmid=row['osmid'], y=row['y'], x=row['x'])

for _, row in G.edges.iterrows():
    graph.run("""
        MATCH (a:Node {osmid: $from}), (b:Node {osmid: $to})
        CREATE (a)-[r:ROAD {length: $length}]->(b)
        RETURN type(r)
    """, from=row['u'], to=row['v'], length=row['length'])

# Exemple de requête Cypher pour trouver le chemin le plus court
start_node = 123
end_node = 456
shortest_path_query = """
MATCH (start:Node {osmid: $start_node}), (end:Node {osmid: $end_node})
CALL gds.alpha.shortestPath.stream({
    nodeProjection: 'Node',
    relationshipProjection: {
        ROAD: {
            type: 'ROAD',
            properties: 'length',
            orientation: 'UNDIRECTED'
        }
    },
    startNode: start,
    endNode: end,
    relationshipWeightProperty: 'length'
})
YIELD nodeId, cost
RETURN gds.util.asNode(nodeId).osmid AS osmid, cost
"""
path_result = graph.run(shortest_path_query, start_node=start_node, end_node=end_node).data()

# Affichage du résultat
for node in path_result:
    print(f"Node OSM ID: {node['osmid']}, Cost: {node['cost']}")
```

```python
import xml.etree.ElementTree as ET
from py2neo import Graph, Node

# Connexion à Neo4j
graph = Graph("bolt://localhost:7687", auth=("neo4j", "123456789"))

# Fonction pour extraire les données du fichier XML et les importer dans Neo4j
def convert_xml_to_neo4j(xml_file):
    # Parsing du fichier XML
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Création d'un nœud pour représenter la ville de Fès
    city_node = Node("City", name="Fès")
    graph.create(city_node)

    # Parcours des éléments XML pour extraire les coordonnées et créer des nœuds Neo4j
    for location in root.findall(".//location"):
        name = location.attrib.get("name", "")
        lat = location.attrib.get("lat", "")
        lon = location.attrib.get("lon", "")

        # Création d'un nœud pour chaque emplacement avec ses coordonnées
        location_node = Node("Location", name=name, latitude=lat, longitude=lon)
        graph.create(location_node)

        # Création d'une relation entre la ville de Fès et l'emplacement
        graph.create((city_node, "LOCATED_IN", location_node))

# Chemin vers le fichier XML contenant les coordonnées de la ville de Fès
xml_file_path = "Fès.xml"

# Appel de la fonction pour convertir le fichier XML en données compatibles avec Neo4j
convert_xml_to_neo4j(xml_file_path)
```
