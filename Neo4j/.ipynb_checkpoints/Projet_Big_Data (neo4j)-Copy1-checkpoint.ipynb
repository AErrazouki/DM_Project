# Téléchargement des données du graphe de la ville de Fès

```python
import osmnx as ox

# Télécharger les données du graphe de la ville de Fès, Maroc
place_name = "Fès, Morocco"
graph = ox.graph_from_place(place_name, network_type="all")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Extraire tous les points d'intérêt (POI) de la ville de Fès
pois = ox.geometries_from_place(place_name, tags={'tourism': 'hotel'})

# Filtrer les géométries pour ne conserver que celles de type Point
hotels_gdf = gpd.GeoDataFrame(pois[pois.geometry.type == 'Point'])

# Afficher la carte de Fès
fig, ax = ox.plot_graph(graph, figsize=(10, 10), show=False, close=False, edge_color='k', node_color='w')

# Ajouter les hôtels à la carte
ax.scatter(hotels_gdf['geometry'].x, hotels_gdf['geometry'].y, color='red', s=50, label='Hôtels')

# Ajouter les noms des hôtels à la carte
for idx, row in hotels_gdf.iterrows():
    ax.text(row['geometry'].x, row['geometry'].y, row['name'], fontsize=8, ha='left', va='center', color='black')

# Afficher la légende et les étiquettes
ax.legend()
plt.title('Hôtels à Fès, Maroc')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python
# Afficher les noms des hôtels sous forme de liste
print("Liste des hôtels à Fès, Maroc:")
for hotel_name in hotels_gdf['name'].unique():
    print("- ", hotel_name)
```

# Stocker les données results dans neo4j

```python
from neo4j import GraphDatabase

# Fonction pour établir une connexion à la base de données Neo4j
def connect_to_neo4j(uri, user, password):
    return GraphDatabase.driver(uri, auth=(user, password))

# Spécifiez les informations de connexion à votre base de données Neo4j
uri = "bolt://localhost:7687"
user = "neo4j"
password = "123456789"

# Établissez une connexion à la base de données Neo4j
driver = connect_to_neo4j(uri, user, password)

# Définissez une fonction pour insérer les données des hôtels dans Neo4j
def insert_hotels_and_location_to_neo4j(hotels_gdf):
    with driver.session() as session:
        # Supprimez d'abord tous les hôtels existants
        session.run("MATCH (h:Hotel) DETACH DELETE h")
         
        # Insérez les nouveaux hôtels
        for idx, row in hotels_gdf.iterrows():
            # Créez une requête Cypher pour insérer un nœud de type "Hotel" avec les propriétés nom et coordonnées
            cypher_query = (
                "CREATE (h:Hotel {name: $name, latitude: $latitude, longitude: $longitude})"
            )
            # Exécutez la requête Cypher avec les paramètres appropriés
            session.run(cypher_query, name=row['name'], latitude=row['geometry'].y, longitude=row['geometry'].x)

            # Créez une relation entre la nouvelle position et les hôtels stockés
            session.run("MATCH (l:Location), (h:Hotel {name: $name}) "
                        "CREATE (l)-[:NEAR_HOTEL]->(h)", name=row['name'])

# Utilisation de la fonction pour insérer les hôtels et la localisation dans Neo4j
insert_hotels_and_location_to_neo4j(hotels_gdf)

print("Les données des hôtels ont été insérées dans Neo4j avec succès.")
```

```python
# Définissez une fonction pour insérer la localisation dans Neo4j
def insert_location_to_neo4j(current_location):
    with driver.session() as session:
        # Supprimez d'abord tous les hôtels existants
        session.run("MATCH (h:Location) DETACH DELETE h")
        # Insérez la localisation dans la base de données
        session.run("CREATE (l:Location {latitude: $latitude, longitude: $longitude})", 
                    latitude=current_location['latitude'], longitude=current_location['longitude'])

# Exemple de localisation actuelle
#current_location = {"latitude": 34.03465334588372, "longitude": -4.975891457672885}
current_location = {"latitude": 34.02100335446571, "longitude": -4.989319457672954}

# Utilisation de la fonction pour insérer la localisation dans Neo4j
insert_location_to_neo4j(current_location)

print("La localisation a été insérée dans Neo4j avec succès.")
```

```python
# Calculer les distances et créer une relation avec l'hôtel le plus proche
def create_relationship_with_closest_hotel(graph, current_location, hotels_gdf):
    with driver.session() as session:
        # Trouver le nœud OSM le plus proche de la localisation actuelle
        current_location_node = ox.distance.nearest_nodes(graph, current_location['longitude'], current_location['latitude'])
        
        closest_hotel_name = None
        closest_hotel_distance = float('inf')
        
        for idx, row in hotels_gdf.iterrows():
            # Trouver le nœud OSM le plus proche de chaque hôtel
            hotel_node = ox.distance.nearest_nodes(graph, row['geometry'].x, row['geometry'].y)
            
            # Calculer la distance ou la durée du trajet entre la localisation actuelle et l'hôtel
            distance = nx.shortest_path_length(graph, current_location_node, hotel_node, weight='length')
            
            if distance < closest_hotel_distance:
                closest_hotel_distance = distance
                closest_hotel_name = row['name']
        
        # Créer la relation avec l'hôtel le plus proche dans Neo4j
        if closest_hotel_name:
            cypher_query = (
                "MATCH (l:Location), (h:Hotel {name: $name}) "
                "CREATE (l)-[:NEAR_HOTEL {distance: $distance}]->(h)"
            )
            session.run(cypher_query, name=closest_hotel_name, distance=closest_hotel_distance)

# Utilisation de la fonction pour créer la relation
create_relationship_with_closest_hotel(graph, current_location, hotels_gdf)

print("La relation avec l'hôtel le plus proche a été créée dans Neo4j avec succès.")
```

### Extraction des donnees 

```python
from neo4j import GraphDatabase

# Fonction pour établir une connexion à la base de données Neo4j
def connect_to_neo4j(uri, user, password):
    return GraphDatabase.driver(uri, auth=(user, password))

# Spécifiez les informations de connexion à votre base de données Neo4j
uri = "bolt://localhost:7687"
user = "neo4j"
password = "123456789"

# Établissez une connexion à la base de données Neo4j
driver = connect_to_neo4j(uri, user, password)

# Définissez une fonction pour extraire les données des hôtels de Neo4j
def extract_hotels_from_neo4j():
    hotels = []
    with driver.session() as session:
        # Exécutez une requête Cypher pour récupérer les noms et coordonnées des hôtels
        result = session.run("MATCH (h:Hotel) RETURN h.name AS name, h.latitude AS latitude, h.longitude AS longitude")
        # Parcourez les résultats et ajoutez chaque hôtel à la liste
        for record in result:
            hotels.append({
                'name': record['name'],
                'latitude': record['latitude'],
                'longitude': record['longitude']
            })
    return hotels

# Récupérez les données des hôtels depuis Neo4j
hotels_from_neo4j = extract_hotels_from_neo4j()
print(hotels_from_neo4j)
```

###  Creation du graphe a partir des donnees extraite 

```python
import networkx as nx
import matplotlib.pyplot as plt
from haversine import haversine

# Your list of hotels
hotels_data = hotels_from_neo4j

# Create a graph
G = nx.Graph()

# Add nodes with positions and names
for hotel in hotels_data:
    if 'name' in hotel and hotel['name'] is not None:
        G.add_node(hotel['name'], pos=(hotel['longitude'], hotel['latitude']))

# Add edges with distances
for hotel1 in hotels_data:
    for hotel2 in hotels_data:
        if hotel1 != hotel2 and 'name' in hotel1 and 'name' in hotel2 and hotel1['name'] is not None and hotel2['name'] is not None:
            distance = haversine((hotel1['latitude'], hotel1['longitude']), (hotel2['latitude'], hotel2['longitude']))
            G.add_edge(hotel1['name'], hotel2['name'], weight=distance)
```
