# Detecting / Treating outliers

```python
df = data.copy()
# Sélectionner uniquement les colonnes numériques
numerical_variables = df.select_dtypes(include=[np.number])

# Calculer les quartiles et l'IQR pour chaque colonne numérique
Q1 = df[numerical_variables.columns].quantile(0.25)
Q3 = df[numerical_variables.columns].quantile(0.75)
IQR = Q3 - Q1

# Calcul des bornes inférieures et supérieures pour chaque colonne numérique
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remplacement des valeurs aberrantes dans chaque colonne numérique
for column in numerical_variables.columns:
    median_value = df[column].median()  # Calcul de la médiane de la colonne
    df[column] = df[column].apply(lambda x: median_value if x < lower_bound[column] or x > upper_bound[column] else x)

# Identifier et afficher le nombre de valeurs aberrantes restantes pour chaque colonne numérique
outliers_count = (df[numerical_variables.columns] < lower_bound) | (df[numerical_variables.columns] > upper_bound)
outliers_count = outliers_count.sum()
print("Nombre de valeurs aberrantes restantes après le remplacement :")
print(outliers_count)
```

# II. Evaluate various machine learning models

# Models

```python
y = df['quality']
X = df.drop(columns = 'quality')
```

La normalisation des données avant l’entraînement aide à obtenir un entraînement stable et rapide du modèle.

```python
# Normalisation des données
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

```python
norm = MinMaxScaler()
# Normaliser les données
X_minmax = norm.fit_transform(X)
```

```python
# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=1)

# Afficher le nombre de lignes et de colonnes de l'ensemble d'entraînement et de l'ensemble de test
print("Dimensions de l'ensemble d'entraînement:")
print(X_train.shape)
print(y_train.shape)

print("\nDimensions de l'ensemble de test:")
print(X_test.shape)
print(y_test.shape)
```

```python
# Initialisation des modèles de classification
models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "Neural Network": MLPClassifier(),
    "KNN": KNeighborsClassifier(),
    "Logistic Regression": LogisticRegression(max_iter=1000)
}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name}: Accuracy = {accuracy:.4f}")
```

```python
# Diviser les données en ensembles d'entraînement et de test
X1_train, X1_test, y1_train, y1_test = train_test_split(X_minmax, y, test_size=0.2, random_state=1)

# Afficher le nombre de lignes et de colonnes de l'ensemble d'entraînement et de l'ensemble de test
print("Dimensions de l'ensemble d'entraînement:")
print(X1_train.shape)
print(y1_train.shape)

print("\nDimensions de l'ensemble de test:")
print(X1_test.shape)
print(y1_test.shape)
```

```python
for name, model in models.items():
    model.fit(X1_train, y1_train)
    y1_pred = model.predict(X1_test)
    accuracy = accuracy_score(y1_test, y1_pred)
    print(f"{name}: Accuracy = {accuracy:.4f}")
```

# Hyperparameter Tuning

 Define hyperparameters and the search space.

 Create a pipeline with feature scaling , StandardScaler() : Normalise les caractéristiques en supprimant la moyenne et en mettant à l'échelle selon l'écart-type.

 Use GridSearchCV et RandomizedSearchCV to find the best hyperparameters .

```python
model = {
    'Decision Tree': (Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', DecisionTreeClassifier())
    ]), {
        'classifier__criterion': ['gini', 'entropy'],
        'classifier__max_depth': [10, 20]
    }),
    'Random Forest': (Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', RandomForestClassifier())
    ]), {
        'classifier__n_estimators': [50, 100],
        'classifier__criterion': ['gini'],
        'classifier__max_depth': [10, 20]
    })
}

# Utilisation de StratifiedKFold pour garantir des proportions de classe équilibrées
cv = StratifiedKFold(n_splits=5)

# Filtrage des avertissements
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=UserWarning)
    
    # Recherche des meilleurs hyperparamètres pour chaque modèle
    best_models = {}
    for name, (pipeline, param_grid) in model.items():
        print(f"Tuning hyperparameters for {name}...")
        grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring='accuracy', n_jobs=-1)
        grid_search.fit(X_train, y_train)
        best_models[name] = grid_search.best_estimator_
        print(f"Best hyperparameters for {name}: {grid_search.best_params_}")
        y_pred = grid_search.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        print(f'Best {name} Accuracy: {accuracy:.4f}')
```

```python
# Filtrage des avertissements
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=UserWarning)
    
    # Recherche des meilleurs hyperparamètres pour chaque modèle avec RandomizedSearchCV
    best_models = {}
    for name, (pipeline, param_distributions) in model.items():
        print(f"Tuning hyperparameters for {name}...")
        random_search = RandomizedSearchCV(
            pipeline, param_distributions, n_iter=10, cv=cv, scoring='accuracy', n_jobs=-1, random_state=42
        )
        random_search.fit(X_train, y_train)
        best_models[name] = random_search.best_estimator_
        print(f"Best hyperparameters for {name}: {random_search.best_params_}")
        y_pred = random_search.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        print(f'Best {name} Accuracy: {accuracy:.4f}')
```

# Cross-Validation

```python
# Définir les métriques de performance à évaluer
scoring = ['precision_macro', 'recall_macro', 'f1_macro', 'roc_auc_ovr', 'accuracy',]
```

```python
# Mettez à jour les modèles en activant la prédiction de probabilité pour SVM
models["SVM"] = SVC(probability=True)

# Effectuer la validation croisée pour chaque modèle
for name, model in models.items():
    print(f"Evaluating model: {name}")
    for metric in scoring:
        scores = cross_val_score(model, X, y, cv=StratifiedKFold(n_splits=5), scoring=metric)
        print(f"Mean {metric}:", scores.mean())
    print()
```

Conclusion :

Le Random Forest semble être le meilleur modèle parmi ceux testés, avec la plus haute AUC ROC et des scores de précision, de rappel et de F1 macro relativement plus élevés. Cependant, les performances générales de tous les modèles semblent être modestes.

# Ensemble Methods

Bagging: Cette méthode forme plusieurs modèles sur différents sous-ensembles de données d'entraînement et combine leurs prédictions par vote majoritaire. Cela aide à réduire la variance des prédictions.

Boosting: Cette méthode ajuste itérativement le poids des observations mal prédites par les modèles précédents. Chaque modèle successif se concentre davantage sur les cas difficiles à prédire, améliorant ainsi les performances globales. Ce modele permet d'optimiser les erreurs résiduelles des prédictions précédentes.

Stacking: Cette méthode combine plusieurs modèles de base pour améliorer la performance globale.. Les prédictions des modèles de base sont utilisées comme caractéristiques d'entrée pour le méta-modèle. Cela permet de capturer des relations plus complexes entre les prédictions des modèles de base.

Voting : Combine plusieurs modèles pour faire des prédictions basées sur un vote majoritaire (hard) ou une moyenne pondérée des probabilités (soft).


```python
# Filtrer les avertissements
warnings.filterwarnings("ignore", message="The least populated class in y has only")
warnings.filterwarnings("ignore", message="lbfgs failed to converge")

# Bagging with RandomForest
print("Evaluating Bagging with RandomForest...")
rf_bagging = RandomForestClassifier(n_estimators=100, random_state=42)
rf_bagging.fit(X_train, y_train)
y_pred_rf_bagging = rf_bagging.predict(X_test)
accuracy_rf_bagging = accuracy_score(y_test, y_pred_rf_bagging)
print(f'RandomForest Bagging Accuracy: {accuracy_rf_bagging:.4f}')

# Boosting with AdaBoost
print("Evaluating Boosting with AdaBoost...")
ada_boost = AdaBoostClassifier(n_estimators=100, random_state=42)
ada_boost.fit(X_train, y_train)
y_pred_ada_boost = ada_boost.predict(X_test)
accuracy_ada_boost = accuracy_score(y_test, y_pred_ada_boost)
print(f'AdaBoost Accuracy: {accuracy_ada_boost:.4f}')

# Boosting with GradientBoosting
print("Evaluating Boosting with GradientBoosting...")
grad_boost = GradientBoostingClassifier(n_estimators=100, random_state=42)
grad_boost.fit(X_train, y_train)
y_pred_grad_boost = grad_boost.predict(X_test)
accuracy_grad_boost = accuracy_score(y_test, y_pred_grad_boost)
print(f'GradientBoosting Accuracy: {accuracy_grad_boost:.4f}')

# Stacking
print("Evaluating Stacking...")
estimators = [
    ('dt', DecisionTreeClassifier()),
    ('rf', RandomForestClassifier()),
    ('svm', SVC(probability=True)),
    ('knn', KNeighborsClassifier())
]
stacking = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
stacking.fit(X_train, y_train)
y_pred_stacking = stacking.predict(X_test)
accuracy_stacking = accuracy_score(y_test, y_pred_stacking)
print(f'Stacking Accuracy: {accuracy_stacking:.4f}')

# Voting (Majority Voting and Soft Voting)
print("Evaluating Voting...")
voting_estimators = [
    ('dt', DecisionTreeClassifier()),
    ('rf', RandomForestClassifier()),
    ('svm', SVC(probability=True)),
    ('knn', KNeighborsClassifier())
]

# Hard Voting
voting_hard = VotingClassifier(estimators=voting_estimators, voting='hard')
voting_hard.fit(X_train, y_train)
y_pred_voting_hard = voting_hard.predict(X_test)
accuracy_voting_hard = accuracy_score(y_test, y_pred_voting_hard)
print(f'Voting (Hard) Accuracy: {accuracy_voting_hard:.4f}')

# Soft Voting
voting_soft = VotingClassifier(estimators=voting_estimators, voting='soft')
voting_soft.fit(X_train, y_train)
y_pred_voting_soft = voting_soft.predict(X_test)
accuracy_voting_soft = accuracy_score(y_test, y_pred_voting_soft)
print(f'Voting (Soft) Accuracy: {accuracy_voting_soft:.4f}')

# Summarize results
ensemble_results = {
    'RandomForest Bagging': accuracy_rf_bagging,
    'AdaBoost': accuracy_ada_boost,
    'GradientBoosting': accuracy_grad_boost,
    'Stacking': accuracy_stacking,
    'Voting (Hard)': accuracy_voting_hard,
    'Voting (Soft)': accuracy_voting_soft
}


print("\nEnsemble Method Accuracies:")
for method_name, accuracy in ensemble_results.items():
    print(f'{method_name}: {accuracy:.4f}')
```

# Conclusion:

RandomForest Bagging: Cette méthode a donné la meilleure précision parmi toutes les méthodes testées, avec une précision de 65%. Cela indique que l'agrégation de multiples modèles RandomForest a bien généralisé sur les données de test.

AdaBoost: AdaBoost a donné la plus faible précision parmi toutes les méthodes évaluées, avec seulement 43%. Cela pourrait indiquer que le modèle AdaBoost a du mal à ajuster les données.

GradientBoosting: GradientBoosting a produit une précision de 57%, ce qui est meilleur que AdaBoost mais moins performant que RandomForest Bagging.

Stacking: Stacking a produit une précision de 64.69%, ce qui est comparable à RandomForest Bagging. Cela indique que la méthode de stacking, qui combine plusieurs modèles de base avec un modèle final, a bien fonctionné dans ce cas.

Voting (Hard): La méthode de vote "hard" a donné une précision de 60.61%, ce qui est inférieure à RandomForest Bagging et Stacking mais supérieure à AdaBoost.

Voting (Soft): La méthode de vote "soft" a produit une précision de 62.65%, ce qui est légèrement inférieur à RandomForest Bagging mais meilleur que Voting (Hard).
